```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('nama.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
salur_red = image_rgb[:, :, 0].astype(int)
salur_green = image_rgb[:, :, 1].astype(int)
salur_blue = image_rgb[:, :, 2].astype(int)
```

```python
plt.figure(figsize=(25, 6))
plt.subplot(2, 4, 1)
plt.imshow(image_rgb)
plt.title('CITRA ASLI')
plt.axis('off')

plt.subplot(2, 4, 2)
plt.hist(salur_red.ravel(), bins=256, range=(0, 256), color='red', alpha=0.5, label='Merah')
plt.hist(salur_green.ravel(), bins=256, range=(0, 256), color='green', alpha=0.5, label='Hijau')
plt.hist(salur_blue.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.5, label='Biru')
plt.title('Histogram')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.legend()
plt.xlim([0, 256])
```

```python
plt.figure(figsize=(25, 6))
plt.subplot(2, 4, 1)
plt.imshow(salur_red, cmap='gray')
plt.title('MERAH')
plt.axis('off')

plt.subplot(2, 4, 2)
plt.hist(salur_red.ravel(), bins=256, range=(0, 256), color='red')
plt.title('Histogram Merah')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.xlim([0, 256])
```

```python
plt.figure(figsize=(25, 6))
plt.subplot(2, 4, 1)
plt.imshow(salur_green, cmap='gray')
plt.title('HIJAU')
plt.axis('off')

plt.subplot(2, 4, 2)
plt.hist(salur_green.ravel(), bins=256, range=(0, 256), color='green')
plt.title('Histogram Hijau')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.xlim([0, 256])
```

```python
plt.figure(figsize=(25, 6))
plt.subplot(2, 4, 1)
plt.imshow(salur_blue, cmap='gray')
plt.title('BIRU')
plt.axis('off')

plt.subplot(2, 4, 2)
plt.hist(salur_blue.ravel(), bins=256, range=(0, 256), color='blue')
plt.title('Histogram Biru')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.xlim([0, 256])
```

```python
gambar = cv2.imread('Qiqi.jpg')
(baris,kolom)=gambar.shape[:2]
```

```python
gray = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)

beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = gray[x,y] + beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)

alpa = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gmx = gray[x,y] * alpa
        citra_kontras[x,y] = gmx

citra_kontras = citra_kontras.astype(np.uint8)

citra_kontras_cerah = np.zeros_like(gambar, dtype=np.uint8)

for x in range(baris):
    for y in range(kolom):
        for c in range(3):  
            nilai = alpha * gambar[x, y, c] + beta
            citra_kontras_cerah[x, y, c] = np.clip(nilai, 0, 255)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(gray)
axs[0,1].imshow(citra_cerah)
axs[1,0].imshow(citra_kontras)
axs[1,1].imshow(citra_kontras_cerah)
```
